```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
c_output = pd.read_csv("c_output.csv", names=["n","execution", "ij", "ji"])
fortran_output = pd.read_csv("fortran_output.csv", names=["n","execution","ij", "ji"])
```

```python
data_c = []
for n in c_output["n"].unique():
  n_df = c_output[ c_output["n"] == n ]
  data_c.append([n, n_df["ij"].mean(),n_df["ji"].mean() ] )
```

```python
c_resumed = pd.DataFrame(data=data_c, columns=["n","ij","ji"])
```

```python
fig = c_resumed.plot(kind="line",x='n', ylabel="time (s)", title="C execution").get_figure()

fig.savefig("c_plot.png")
```

```python
data_fortran = []
for n in fortran_output["n"].unique():
  n_df = fortran_output[ fortran_output["n"] == n ]
  data_fortran.append([n, n_df["ij"].mean(),n_df["ji"].mean() ] )
```

```python
fortran_resumed = pd.DataFrame(data=data_fortran, columns=["n","ij","ji",])
```

```python
fig = fortran_resumed[["n","ij", "ji"]].plot(kind="line",x='n', ylabel="time (s)", title="Fortran execution").get_figure()

fig.savefig("fortran_plot.png")
```
